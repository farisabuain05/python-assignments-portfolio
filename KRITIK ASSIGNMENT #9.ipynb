```python
## EXERCISE 2: Given the function g(x, y) = (x^2)y + x(y^2), find the gradient vector and compute its magnitude 
## at the point (1, −1).

g = (x**2)*y + x*(y**2)

dg_dx = sp.diff(g, x)

dg_dy = sp.diff(g, y)

## gradient vector is ∇g = (dg/dx, dg/dy)

## the magnitude of ∇g at (1, -1) is sqrt(dg/dx(1, -1)^2, dg/dy(1, -1)^2)

dg_dx_1 = dg_dx.subs({x:1, y:-1}) # partial w.r.t. x evaluated at the point (1,-1)

p_1_dg_dx = dg_dx_1**2 # square for magnitude summation

dg_dy_1 = dg_dy.subs({x:1, y:-1}) # partial w.r.t. y evaluated at the point (1,-1)

p_1_dg_dy = dg_dy_1**2 # square for magnitude summation

mag_gradient = sp.sqrt(p_1_dg_dx + p_1_dg_dy) # sum the squares and square root the sum to find magnitude!

print("The magnitude of the gradient vector at (1, -1) is", mag_gradient)
```

```python
## EXERCISE 3: For the function h(x, y) = ln(x^2 +y^2), find the second partial derivatives ∂^2h/∂x^2 , ∂^2h/∂y^2 , 
## and ∂^2h/∂x∂y . Discuss the symmetry of mixed partial derivatives.

h = log(x**2 + y**2)

dh_dx = sp.diff(h, x)

dh_dy = sp.diff(h, y)

## Recall:
## d^2h/dx^2 = d/dx(dh/dx)
## d^2h/dy^2 = d/dy(dh/dy)
## d^2h/dxdy = d/dx(dh/dy) OR d/dy(dh/dy)
## remember, mixed partials are identical for well-behaved functions. Since log(x**2 + y**2) is a composite function
## of two integrable and differentiable functions, we can be reasonably sure that it is well-behaved.

d2h_dx2 = sp.diff(dh_dx, x)

d2h_dy2 = sp.diff(dh_dy, y)

d2h_dxdy = sp.diff(dh_dy, x)

print(d2h_dx2)
print(d2h_dy2)
print(d2h_dxdy)

## to prove the symmetry of mixed partials, let's find the other way of mixed partial and equate them:

d2h_dydx = sp.diff(dh_dx, y)

print("As we can see", d2h_dydx, "=", d2h_dxdy, ", thus the mixed partials are equivalent for this particular function")
```

```python
## EXERCISE 4: Create a contour plot of the function j(x, y) = x^3 − 3xy + y^3 . 

## Step 1: Import the necessary libraries

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

## Step 2: Define the function using SymPy

x, y = sp.symbols('x y')

j = x**3 - 3*x*y + y**3

## Step 3: Convert the SymPy expression to a numerical function

j_func = sp.lambdify((x, y), j, 'numpy')

## Step 4: Create a grid of x and y values, and evaluate j on this grid

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)

X, Y = np.meshgrid(x_vals, y_vals)

Z = j_func(X, Y)

## Step 5: Plot the contour

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
